Matplotlib is a fantastic library for plotting and visualization of results and data. It is capable of displays in some many formats. In this notbook however we will only exlpore the very basics and as we explore the AI more there will be several instances where some advanced features shall be put to use.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
#This is MATLAB style plotting.
plt.plot(x, np.sin(x), 'b-', label='sine')
plt.plot(x, np.cos(x), 'g-', label='cosine')
plt.xlabel('Angle')
plt.ylabel('Sine/Cosine')
plt.title('Sinusoids')
plt.xlim(0, 10)
plt.ylim(-1,1)
plt.legend(loc = 'upper right')
plt.show()
```

```python
#Object oriented matplotlib. More python way.
fig = plt.figure()
#Sub-plot: In same figure with 2 rows, 1 column, 1-position of current plot.
ax1 = fig.add_subplot(2,1,1)
ax1.plot(x, np.sin(x), 'r-', label='sine')
ax1.set_xlim(0,10)
ax1.set_ylim(-1,1)
#Sub-plot: In same figure with 2 rows, 1 column, 2-position of current plot.
ax1.legend(loc='upper right')
ax2 = fig.add_subplot(2,1,2)
ax2.plot(x, np.cos(x), 'g-', label='cosine')
ax2.set_xlim(0,10)
ax2.set_ylim(-1,1)
ax2.legend(loc='upper right')
plt.show()
```

Plotting zipped data.If data is already available then how to unzip and plot it? Read-on.

```python
x_data = np.arange(0,5)
y_data = x_data ** 2
x_data, y_data
```

```python
zipped_data = list(zip(x_data, y_data))
zipped_data
```

```python
plt.scatter(*zip(*zipped_data))    
plt.show()
```

Meshgrid: A tutorial on meshgrid. With data x of shape (Sx,) and data y of shape (Sy,), meshgrid assumes shapes for x and y as (Sx, Sy).

```python
grid1 = np.arange(10,20,1)
grid2 = np.arange(20,25,1)
grid1, grid2
```

```python
x_grid1, y_grid2 = np.meshgrid(grid1, grid2)
x_grid1, y_grid2
```
